```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../Data/cleaned_data.csv')
df['month'] = pd.to_datetime(df['month'])
```

```python
monthly_sales = df.groupby('month')['sales'].sum()

monthly_sales.plot(figsize=(10,5), title='Sales Trend Over Time')
plt.show()
```

```python
sns.scatterplot(x=df['strategy1'], y=df['sales'])
plt.title('Strategy1 vs Sales')
plt.show()
```

```python
sns.scatterplot(x=df['strategy2'], y=df['sales'])
plt.title('Strategy2 vs Sales')
plt.show()
```

```python
sns.scatterplot(x=df['strategy3'], y=df['sales'])
plt.title('Strategy3 vs Sales')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
```

```python
sns.barplot(x='acctype', y='sales', data=df)
plt.xticks(rotation=45)
plt.title('Sales by Account Type')
plt.show()
```

```python
df['post_competitor'] = df['month'] >= '2015-06-01'
```

```python
sns.boxplot(x='post_competitor', y='sales', data=df)
plt.title('Sales Before vs After Competitor')
plt.show()
```

```python
df.groupby('post_competitor')['sales'].mean()
```

```python
before = df[df['post_competitor'] == False]['sales'].mean()
after = df[df['post_competitor'] == True]['sales'].mean()

print("Before:", before)
print("After:", after)
```

```python
sns.barplot(x='district', y='sales', data=df)
plt.title('Sales by District')
plt.show()
```
